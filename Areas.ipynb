# Áreas de jardinagem

# Lista de Areas cadastradas

```python
# retorna todas as áreas cadastradas, que o usário tem permissão para visualizar
# codigo de permissão necessário --> 252: Pode visualizar áreas de jardinagem
response = requests.get("http://127.0.0.1:9000/api/areas/ListAreasJardinagem/", headers=headers)
response.json()['results']
df = pd.DataFrame(response.json()['results'])
display(df)
```

# Criar nova área

```python
## As colunas, Terreno, vegetacao, servico, localidade são chaves estrangeiras e precisam estar devidamente configuradas
## sendo obriagatorios
```

```python
# Lista de terenos que podem ser enviados pelo usuario logado
# nao precisa de permissao
print("Terrenso disponiveis")
response = requests.get("http://127.0.0.1:9000/api/terrenos/ListTerrenosFromForms/", headers=headers)
response.json()
df = pd.DataFrame(response.json()['results'])
display(df)
print()


# Lista de vegetacoes que podem ser enviados pelo usuario logado
# nao precisa de permissao
print("Vegetações disponiveis")
response = requests.get("http://127.0.0.1:9000/api/vegetacao/ListCatalogoVegetacaoFromForms/", headers=headers)
response.json()
df = pd.DataFrame(response.json()['results'])
display(df)


# Lista de servicos que podem ser enviados pelo usuario logado
# nao precisa de permissao
print("Servicos disponiveis")
response = requests.get("http://127.0.0.1:9000/api/catalogo_de_servicos/ListCatalogodeServicoJardinagemFromForms/", headers=headers)
response.json()
df = pd.DataFrame(response.json()['results'])
display(df)


# Lista de localidades que podem ser enviados pelo usuario logado
# nao precisa de permissao
print("Localidades disponiveis")
response = requests.get("http://127.0.0.1:9000/api/localidades/ListLocalidadeJardinagemFromForms/", headers=headers)
response.json()
df = pd.DataFrame(response.json()['results'])
display(df)
```

# Atualizar area

```python
# as chaves body da atualização é exatamente igual o body da criacao
# atualiando a periodicidade para bimestral
body = {
    'nome': "area jupyter", 
    'dimensao': 1568, 
    'Terreno': 5, # vem de api/terrenos/ListTerrenosFromForms/
    'vegetacao': 2, # vem de api/vegetacao/ListCatalogoVegetacaoFromForms/
    'servico': 2, # vem de /api/catalogo_de_servicos/ListCatalogodeServicoJardinagemFromForms/
    'localidade': 4,  # vem de api/localidades/ListLocalidadeJardinagemFromForms/
    'foto': "", #Tipo file    
    'periodicidade': "Bimestral", # opcoes disponiveis 'Mensal', 'Semanal', 'Quinzenal', 'Bimestral', 'Trimestral', 'Semestral', 'Anual',
}
```

```python
# essa rota recebe o id random da área como parametro de url, consultar em api/areas/ListAreasJardinagem/
# atualizando a área criada anteriormente, obeserve o id_random
# codigo de permissão necessário --> 251: Pode editar áreas de jardinagem
response = requests.put("http://127.0.0.1:9000/api/areas/AreasJardinagemUpdate/IEWk9geSLqsF/", body, headers=headers)
response.json()
```

# Alterar status area

```python
# essa rota recebe o id random da área como parametro de url, consultar em api/areas/ListAreasJardinagem/
# atualizando a área criada anteriormente, obeserve o id_random
# codigo de permissão necessário --> 
# 254: Pode desmobilizar áreas de jardinagem ou demosbilizar (Desmobilizado) ou
# 255: Pode reabilitar áreas de jardinagem para reabilitar (Mobilizado)

body = {
    'status': 'Desmobilizado',
}

response = requests.put("http://127.0.0.1:9000/api/areas/AreasJardinagemAlterStatus/IEWk9geSLqsF/", body, headers=headers)
response.json()
```
