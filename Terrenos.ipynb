# Atualizar terrenos

```python
# as chaves body da atualização é exatamente igual o body da criacao
# atualiando o nome para 'terreno edita jupyter
new_body = {
     'nome': 'terreno edita jupyter',
     'EmpresaSecundaria': 6,  # vem de api/empresas_secundarias/ListEmpresaSecundariaJardinagemFromForms/
}

# essa rota recebe o id random da tereno como parametro de url, consultar em api/terrenos/ListTerrenos/
# atualizando a servico criada anteriormente, obeserve o id_random
# codigo de permissão necessário --> 331: Pode editar terrenos
response = requests.put("http://127.0.0.1:9000/api/terrenos/TerrenoUpdate/PVtOuGGY36VU/", new_body, headers=headers)
response.json()
```

# Detalhes terrenos

```python
# essa rota recebe o id random da terreno como parametro de url, consultar em api/terrenos/ListTerrenos/
# detalhes do terreno criado anteriormente, obeserve o id_random
# http://127.0.0.1:9000/api/terrenos/TerrenoDetail/id_random/
# codigo de permissão necessário --> 332: Pode visualizar terrenos
response = requests.get("http://127.0.0.1:9000/api/terrenos/TerrenoDetail/PVtOuGGY36VU/", new_body, headers=headers)
response.json()
```

# Alterar status

```python
# essa rota recebe o id random da terenos como parametro de url, consultar em api/terrenos/ListTerrenos/
# atualizando a terreno criada anteriormente, obeserve o id_random
# codigo de permissão necessário --> 
# 334: Pode desmobilizar terrenos para demosbilizar (Desmobilizado) ou
# 335: Pode reabilitar terrenos para reabilitar (Mobilizado)
new_body = {
    'status': 'Desmobilizado'
}

response = requests.put("http://127.0.0.1:9000/api/terrenos/TerrenoAlterStatus/PVtOuGGY36VU/", new_body, headers=headers)
response.json()
```

# If delete Terreno

```python
# essa rota recebe o id random da terreno como parametro de url, consultar em api/terrenos/ListTerrenos/
# codigo de permissão necessário --> 333: Pode excluir terrenos
response = requests.get("http://127.0.0.1:9000/api/terrenos/IfDeleteTerreno/PVtOuGGY36VU/", new_body, headers=headers)
response.json()
```

# Deleção de terrenos

```python
# essa rota recebe o id random da terreno como parametro de url, consultar em api/terrenos/ListTerrenos/
# codigo de permissão necessário --> 333: Pode excluir terrenos
response = requests.delete("http://127.0.0.1:9000/api/terrenos/DeleteTerreno/PVtOuGGY36VU/", headers=headers)
response.json()
```
