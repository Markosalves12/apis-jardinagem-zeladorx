# Criar novo servicos configurado

```python
## As colunas, Areas, ServicosEscalados, diasaseremrealizado localidade são chaves estrangeiras e precisam estar devidamente configuradas
## sendo obriagatorios
```

```python
# lista de areas dispniveis para solitarem acesso
response = requests.get("http://127.0.0.1:9000/api/areas/ListAreasJardinagemFromFoms/", headers=headers)
response.json()['results']
df = pd.DataFrame(response.json()['results'])
display(df)
print()


# servicos disponiveis na configuracao
response = requests.get("http://127.0.0.1:9000/api/catalogo_de_servicos/ListCatalogodeServicoJardinagemFromForms/", headers=headers)
response.json()['results']
df = pd.DataFrame(response.json()['results'])
display(df)
print()


# Dias da semana disponiveis
response = requests.get("http://127.0.0.1:9000/api/semana/ListDiasDaSemana/", headers=headers)
response.json()['results']
df = pd.DataFrame(response.json()['results'])
display(df)
print()
```

```python
# codigo de permissão necessário --> 370: Pode configurar novos serviços
response = requests.post("http://127.0.0.1:9000/api/servicos/configurados/CreateServicosConfiguradosJardins/", body, headers=headers)
response.json()
```

# Alterar Status do servico configurado

```python
# essa rota recebe o id random da servico configurado como parametro de url, consultar em api/servicos/configurados/ListServicoJardinagemConfigurado/
# atualizando a área criada anteriormente, obeserve o id_random
# codigo de permissão necessário --> 
# 374: Pode desmobilizar serviços configurados para demosbilizar (Desmobilizado) ou
# 375: Pode reabilitar serviços configurados para reabilitar (Mobilizado)

body = {
    'status': 'Desmobilizado'
}

response = requests.put("http://127.0.0.1:9000/api/servicos/configurados/ServicosConfiguradosJardinsAlterStatus/efwDKeekrM9I/", body, headers=headers)
response.json()
```
