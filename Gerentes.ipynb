```python
# Lista de empresas que podem ser enviadas pelo usuario
# nao precisa de permissao
print("Empresas secundarias")
response = requests.get("http://127.0.0.1:9000/api/empresas_secundarias/ListEmpresaSecundariaJardinagemFromForms/", headers=headers)
response.json()
df = pd.DataFrame(response.json()['results'])
display(df)
print()
```

```python
# cria um usuario especifico
# o campo empresasecundaria é chave estrqangeira e precisa esta configrado
# codigo de permissão necessário --> 280: Pode criar novos colaboradores
body = {
    'username': "user jupyter", 
    'email': "userjupyter@lmail.com", 
    'empresasecundaria': 7, # vem de # api/empresas_secundarias/ListEmpresaSecundariaJardinagemFromForms/
    'is_superuser': 'true', #true ou false, esse campo so pode alterado por outro superusuario 
    'status': "Mobilizado", #pode ser Mobilziado ou Desmobilizado
}

response = requests.post("http://127.0.0.1:9000/api/gerentes/CreateGerenteJardinagem/", body, headers=headers)
response.json()
```

# atualizar gerente

```python
# as chaves body da atualização é exatamente igual o body da criacao
# atualiando a username para user jupyter update
body = {
    'username': "user jupyter update", 
    'email': "userjupyter@lmail.com", 
    'empresasecundaria': 7, # id random da empresa secundaria disponiel em ListEmpresaSecundariaJardinagemFromForms
    'is_superuser': 'false', #true ou false, esse campo so pode alterado por outro superusuario 
}
```

```python
# essa rota recebe o id random da gerente como parametro de url, consultar em api/gerentes/ListGerenteJardinagem/
# atualizando a gerente criada anteriormente, obeserve o id_random
# codigo de permissão necessário --> 281: Pode editar colaboradores
response = requests.put("http://127.0.0.1:9000/api/gerentes/GerenteJardinagemUpdate/CfzvMuv4wClf/", body, headers=headers)
response.json()
```

# Alterar Status do gernete

```python
# essa rota recebe o id random da gerente como parametro de url, consultar em api/gerentes/ListGerenteJardinagem/
# atualizando a gerente criada anteriormente, obeserve o id_random
# codigo de permissão necessário --> 
# 284: Pode desmobilizar colaboradores para demosbilizar (Desmobilizado) ou
# 285: Pode reabilitar colaboradores (Mobilizado)

body = {
    'status': "Desmobilizado"
}

response = requests.put("http://127.0.0.1:9000/api/gerentes/GerenteJardinagemAlterStatus/CfzvMuv4wClf/", body, headers=headers)
response.json()
```
