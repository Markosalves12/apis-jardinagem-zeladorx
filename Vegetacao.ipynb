# Detalhes vegetação

```python
# essa rota recebe o id random da vegetacao como parametro de url, consultar em api/vegetacao/ListCatalogoVegetacao/
# detalhes do terreno criado anteriormente, obeserve o id_random
# http://127.0.0.1:9000/api/vegetacao/VegetacaoDetail/id_random/
# codigo de permissão necessário --> 352: Pode visualizar vegetações

response = requests.get("http://127.0.0.1:9000/api/vegetacao/VegetacaoDetail/biiFu1kF6WGq/", headers=headers)
response.json()
```

# atualizar vegetação

```python
# as chaves body da atualização é exatamente igual o body da criacao
# atualiando o nome para 'vegeta jupyter
body = {
    'nome': 'vegeta jupyter',
    'EmpresaSecundaria': 9,  # vem de api/empresas_secundarias/ListEmpresaSecundariaJardinagemFromForms/
}
# essa rota recebe o id random da vegetacao como parametro de url, consultar em api/vegetacao/ListCatalogoVegetacao/
# atualizando a vegetacao criada anteriormente, obeserve o id_random
# codigo de permissão necessário --> 351: Pode editar vegetações

response = requests.put("http://127.0.0.1:9000/api/vegetacao/VegetacaoUpdate/UA7UolgXJG0I/", body, headers=headers)
response.json()
```

# Alterar status vegetação

```python
# essa rota recebe o id random da vegetacao como parametro de url, consultar em api/vegetacao/ListCatalogoVegetacao/
# atualizando a terreno criada anteriormente, obeserve o id_random
# codigo de permissão necessário --> 
# 354: Pode desmobilizar vegetações para demosbilizar (Desmobilizado) ou
# 355: Pode reabilitar vegetações para reabilitar (Mobilizado)

body = {
    'status': 'Desmobilizado',
}

response = requests.put("http://127.0.0.1:9000/api/vegetacao/VegetacaoAlterStatus/UA7UolgXJG0I/", body, headers=headers)
response.json()
```

# Se deletar vegetação

```python
# essa rota recebe o id random da vegetacao como parametro de url, consultar em api/vegetacao/ListCatalogoVegetacao/
# codigo de permissão necessário --> 353: Pode excluir vegetações
response = requests.get("http://127.0.0.1:9000/api/vegetacao/IfDeleteCatalogoVegetacao/UA7UolgXJG0I/", headers=headers)
response.json()
```

# delecao da vegetação

```python
# essa rota recebe o id random da vegetacao como parametro de url, consultar em api/vegetacao/ListCatalogoVegetacao/
# codigo de permissão necessário --> 353: Pode excluir vegetações
response = requests.delete("http://127.0.0.1:9000/api/vegetacao/DeleteCatalogoVegetacao/biiFu1kF6WGq/", headers=headers)
response.json()
```
